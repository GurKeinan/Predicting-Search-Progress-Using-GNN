```python
import pandas as pd
import pathlib
import matplotlib.pyplot as plt
```

```python
root_dir = pathlib.Path().resolve().parent
csv_output_dir = root_dir / 'csv_output'
```

```python
number_of_nodes_list = []
avg_branching_factor_list = []
max_branching_factor_list = []
solution_length_list = []

for file in csv_output_dir.glob('*.csv'):
    df = pd.read_csv(file)
    number_of_nodes_list.append(len(df))
    avg_branching_factor_list.append(df['BF'].mean())
    max_branching_factor_list.append(df['BF'].max())
    solution_length_list.append(df['g'].iloc[-1])
```

```python
# number of nodes histogram
plt.hist(number_of_nodes_list, bins=50)
plt.title('Number of nodes histogram')
plt.xlabel('Number of nodes')
plt.ylabel('Frequency')
plt.show()

# avg branching factor histogram
plt.hist(avg_branching_factor_list, bins=50)
plt.title('Average branching factor histogram')
plt.xlabel('Average branching factor')
plt.ylabel('Frequency')
plt.show()

# max branching factor histogram
plt.hist(max_branching_factor_list, bins=50)
plt.title('Max branching factor histogram')
plt.xlabel('Max branching factor')
plt.ylabel('Frequency')
plt.show()

# solution length histogram
plt.hist(solution_length_list, bins=50)
plt.title('Solution length histogram')
plt.xlabel('Solution length')
plt.ylabel('Frequency')
plt.show()
```
